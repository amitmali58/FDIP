```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python

img = imread("img3")
imshow(img);
img.shape
```

```python
fig, ax = plt.subplots(1, 3,
figsize=(6,4),
sharey= True)

ax[0].imshow(img[:, 0:341])
ax[0].set_title('First Split')

ax[1].imshow(img[:, 341:682])
ax[1].set_title('Second Split')

ax[2].imshow(img[:, 682:1024])
ax[2].set_title('Third Split');
```

```python
imshow(img[350:600, 100:900]);
```

```python
fig, ax = plt.subplots(1, 3,
figsize=(6,4),
sharey= True)

ax[0].imshow(img[0:220, :])
ax[0].set_title('First Split')

ax[1].imshow(img[220:450, :])
ax[1].set_title('Second Split')

ax[2].imshow(img[450:682, :])
ax[2].set_title('Third Split');
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(img[:,:,0], cmap='Reds')
ax[0].set_title('Red')
ax[1].imshow(img[:,:,1],cmap='Greens')
ax[1].set_title('Green')
ax[2].imshow(img[:,:,2],cmap='Blues')
ax[2].set_title('Blue');
```

```python
img_hsv = rgb2hsv(img)
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(img_hsv[:,:,0], cmap='hsv')
ax[0].set_title('Hue')
ax[1].imshow(img_hsv[:,:,1],cmap='gray')
ax[1].set_title('Saturation')
ax[2].imshow(img_hsv[:,:,2], cmap='gray')
ax[2].set_title('Value');
```

```python
img_gray = rgb2gray(img)
fig,ax =plt.subplots(1, 5, figsize=(17,6), sharey = True)
ax[0].imshow(img_gray, cmap = 'gray')
ax[0].set_title('Grayscale Original')
ax[1].imshow(img_as_uint(img_gray  >0.25),cmap = 'gray')
ax[1].set_title('Greater than 0.25')
ax[2].imshow(img_as_uint(img_gray  >0.50),cmap = 'gray')
ax[2].set_title('Greater than 0.50');
ax[3].imshow(img_as_uint(img_gray  > 0.75),cmap = 'gray')
ax[3].set_title('Greater than 0.75');
ax[4].imshow(img_as_uint(img_gray  > np.mean(img_gray )),cmap = 'gray')
ax[4].set_title('Greater than Mean');
```
