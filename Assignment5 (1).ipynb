```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_color = cv2.imread(r"img1.jpg")

img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title('Original Color Image', fontsize=14)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Grayscale Image', fontsize=14)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(img_gray.ravel(), 256, [0, 256], color='gray')
plt.title('Grayscale Histogram', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
min_val = np.min(img_gray)
max_val = np.max(img_gray)
contrast_stretched = ((img_gray - min_val) / (max_val - min_val) * 255).astype(np.uint8)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(contrast_stretched, cmap='gray')
plt.title('Contrast Stretched Image', fontsize=14)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(contrast_stretched.ravel(), 256, [0, 256], color='gray')
plt.title('Contrast Stretched Histogram', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
equalized = cv2.equalizeHist(img_gray)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(equalized, cmap='gray')
plt.title('Histogram Equalized Image', fontsize=14)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(equalized.ravel(), 256, [0, 256], color='gray')
plt.title('Equalized Histogram', fontsize=14)

plt.tight_layout()
plt.show()
```
