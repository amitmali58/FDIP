```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt 
from skimage.io import imshow, imread

img = imread("assi1img.jpg")
imshow(img)
```

```python
grey=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
imshow(grey)
```

```python
bright_inc = np.clip(grey + 50, 0, 255).astype(np.uint8)
imshow(bright_inc)
```

```python
bright_dec = np.clip(grey - 50, 0, 255).astype(np.uint8)
imshow(bright_dec)
```

```python
neg = 255-grey
imshow(neg)
```

```python
_, thresh = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
imshow(thresh)
```

```python
c = 255 / np.log(1 + np.max(grey))
log_img = c * (np.log(1 + grey))
log_img = np.array(log_img, dtype=np.uint8)
imshow(log_img)
```

```python
norm = grey / 255.0
gamma = 0.5  
gamma_img = np.power(norm, gamma)
gamma_img = np.uint8(gamma_img * 255)
imshow(gamma_img)
```
